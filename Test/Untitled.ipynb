```python
df1=pd.read_csv("filteredCountry.csv")
df1
```

```python
num=df1['SKU'].unique()
num
```

```python
df1['PRICE']=df1['PRICE'].str.replace("$","")
df1['PRICE']
```

```python
first=list(df1["PRICE"].groupby(df1['SKU']))
firstlist=[]
for i in range(len(first)):
    temp=first[i][1].array
    firstlist.append(temp[0])  
```

```python
firstlist
```

```python
li=list(df1["PRICE"].groupby(df1['SKU']))
```

```python
second=[]
for i in range(len(li)):
    if(len(li[i][1])==1):
        temp=li[i][1].array  
        second.append(temp[0])
    elif(len(li[i][1])>1):       
        temp=li[i][1].array
        second.append(temp[1])   
        
```

```python

len(second)
            
```

```python
second
```

```python
result=[num,firstlist,second]
```

```python
df3 = pd.DataFrame(result, index = ['SKU', 'FIRST_MINIMUM_PRICE','SECOND_MINIMUM_PRICE'])  
```

```python
df3
```

```python
df1_transposed = df3.T # or df1.transpose()
df1_transposed
```

```python
df1_transposed.to_csv('output/lowestPrice.csv',index=False,header=True) 
```
